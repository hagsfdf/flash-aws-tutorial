```python
import urllib.request
from bs4 import BeautifulSoup

def save_txt(url):
    page = urllib.request.urlopen(url)
    soup = BeautifulSoup(page, "lxml")

    imglink = soup.find('div', id='titleStoryLine')

    aa = imglink.find('div', class_='inline canwrap')

    t = aa.find('span').get_text().strip()
    return t
```
